```python
import os
import pickle
import matplotlib
from tqdm import tqdm
import csv
from itertools import islice
from pathlib import Path
from typing import Any, Callable, Dict, List, Optional, Tuple, Union

import numpy as np

import torch
import torch.nn as nn
from torchvision import datasets
from torchvision import transforms
from torch.utils.data.sampler import SubsetRandomSampler
from torch.utils.data import DataLoader
from torchvision.datasets import VisionDataset
from torchvision.transforms import v2
from torchsummary import summary
from torchvision.transforms.functional import pil_to_tensor
from torchvision.transforms.functional import to_pil_image

from PIL import Image

import matplotlib.pyplot as plt

# todo:
#
# inference mode: use a model to classify an image, show prediction and img (and emoji)
# try other models (deconvolution for visualization)
#
# pyplot with multiple images, label and predictions
# choose the right transforms, similar to fer+ paper (or simply slight rotate/traslate, skew, crop..)
#
# normalize dataset: https://discuss.pytorch.org/t/computing-the-mean-and-std-of-dataset/34949/17
# (check how normalizatin in fer+)
# both batch normalization and dataset-wide normalization, check slides on this, and where/if to use it in the model training..
#
# save model, reuse to show predictions on test set / val set
# save checkpoints
# save learning curve data to csv


# visualize img transforms, also for the fer+ original code
# check transforms: that are actually been used (visualize images from dataloader); and that the transform is different at each epoch (visualize same img on different loops
#
# plot the hps, save the plot to file at each refresh, also to ..csv to later draw multilpe learning curves
#
# (show learning curve..)
#
# (initialization? seed for reproducibility?)
#
# ? remove NF, unknown classes?
#
# compute other metrics that balance the classes
#

#
# count params and shapes of layers and overall model
# print model summary
#
# ? 64×64 resolution ? (check their code ..)
# https://github.com/Microsoft/FERPlus
#
# hps: lr, ..
#
# ? discard last minibatch because smaller than batch size? (unbalanced influence on training)
# done:
# print model summary, shape of layers, param counts
# rerun training with new labels, new hps (lr=0.05, batch_size=32, ..)
# print stats of dataset (count per class)
# convert merge datasets FER, FERplus (majority voting)
#
# do the mapping to the NER+ labels
# (error checking: percentace of fer+ labels by majority that are the same as original FER)
# FER original: "emotion,pixels,Usage"
# FER+: "The format of the CSV file is as follows:
# usage, neutral, happiness, surprise, sadness, anger, disgust, fear, contempt, unknown, NF.
# Usage,Image name,neutral,happiness,surprise,sadness,anger,disgust,fear,contempt,unknown,NF
# Columns "usage" is the same as the original FER label to differentiate between training, public test and private test sets.
# The other columns are the vote count for each emotion with the addition of unknown and NF (Not a Face)."
#
# "extract the images from the FER csv file in such a way, that
# all images corresponding to "Training" go to FER2013Train folder, all images corresponding to "PublicTest" go to FER2013Valid folder
# and all images corresponding to "PrivateTest" go to FER2013Test folder.
# Or you can use generate_training_data.py script to do all the above for you as mentioned in next section."
#
# cnn FER

# increase accuracy of current NN (try to repro paper at 83%)
# then switch to a NN with visualizations (deconvolutions)

# ..

# id2emo
FER2013Emotion_decode: dict = {
    0:"Angry", 1:"Disgust", 2:"Fear", 3:"Happy", 4:"Sad", 5:"Surprise", 6:"Neutral"
}

# emo2id
emotion_table_FERPlus = {'neutral'  : 0,
                 'happiness': 1,
                 'surprise' : 2,
                 'sadness'  : 3,
                 'anger'    : 4,
                 'disgust'  : 5,
                 'fear'     : 6,
                 'contempt' : 7}
FERP_COL_FILENAME = 1
FERO_COL_EMO = 0
FERO_COL_IMG = 1
FERO_COL_USAGE = 2
FERP_COL_USAGE = 0

fer_orig_emotion_labels = ["Angry",  "Disgust",  "Fear",  "Happy",  "Sad",  "Surprise",  "Neutral"]
ferplus_emotion_labels = ["neutral","happiness","surprise","sadness","anger","disgust","fear","contempt","unknown","NF"]
ferplus_csv_cols = ["Usage","Image name"]
ferplus_csv_cols.extend(ferplus_emotion_labels)
fer_orig_cols = ["emotion","pixels","Usage"]



def get_label2id_id2label(labels: List[str])-> Tuple[Dict[int, str], Dict[str, int]]:

    id2label: Dict[int, str] = dict()
    label2id: Dict[str, int] = dict()
    for id, label in enumerate(labels):
        label2id[label] = id
        id2label[id] = label

    return id2label, label2id

def save2pickle(filepath: str, my_obj):
        filepath = os.path.abspath(filepath)
        # todo: save to pickle
        with open(filepath, 'wb') as file:
            # A new file will be created
            pickle.dump(my_obj, file)
        print(f"Obj saved to pickle file with {filepath=}")

def load_from_pickle(filepath: str):
    filepath = os.path.abspath(filepath)
    with open(filepath, 'rb') as file:
        my_obj = pickle.load(file)
        print(f"Obj loaded from pickle file with {filepath=}")
        return my_obj

# def data_loader_cifar(data_dir,
#                 batch_size,
#                 random_seed=42,
#                 valid_size=0.1,
#                 shuffle=True,
#                 test=False):

#     normalize = transforms.Normalize(
#         mean=[0.4914, 0.4822, 0.4465],
#         std=[0.2023, 0.1994, 0.2010],
#     )

#     # define transforms
#     transform = transforms.Compose([
#             transforms.Resize((227,227)),
#             transforms.ToTensor(),
#             normalize,
#     ])

#     if test:
#         dataset = datasets.CIFAR100(
#           root=data_dir, train=False,
#           download=True, transform=transform,
#         )

#         data_loader: DataLoader = DataLoader(
#             dataset, batch_size=batch_size, shuffle=shuffle
#         )

#         return data_loader

#     # load the dataset
#     train_dataset = datasets.CIFAR100(
#         root=data_dir, train=True,
#         download=True, transform=transform,
#     )

#     valid_dataset = datasets.CIFAR10(
#         root=data_dir, train=True,
#         download=True, transform=transform,
#     )

#     num_train = len(train_dataset)
#     indices = list(range(num_train))
#     split = int(np.floor(valid_size * num_train))

#     if shuffle:
#         np.random.seed(random_seed)
#         np.random.shuffle(indices)

#     train_idx, valid_idx = indices[split:], indices[:split]
#     train_sampler = SubsetRandomSampler(train_idx)
#     valid_sampler = SubsetRandomSampler(valid_idx)

#     train_loader: DataLoader = DataLoader(
#         train_dataset, batch_size=batch_size, sampler=train_sampler)

#     valid_loader: DataLoader = DataLoader(
#         valid_dataset, batch_size=batch_size, sampler=valid_sampler)

#     return (train_loader, valid_loader)

# def data_loader_FER_Plus(data_dir,
#                 batch_size,
#                 random_seed=42,
#                 valid_size=0.1,
#                 shuffle=True,
#                 test=False):
#     pass

def merge_fer_ferplus(
        fer_orig_csv_path,
        ferplus_csv_path,
        out_merged_csv_path,
):
    # todo:
    # merge the 2 datasets
    # check statistic on how many have same label
    # also save to disk like pickle to avoid processing every time

    # FER original: "emotion,pixels,Usage"
    # FER+: "The format of the CSV file is as follows:
    # usage, neutral, happiness, surprise, sadness, anger, disgust, fear, contempt, unknown, NF.
    # merged format 1:
    # keep the format (cols) of fer orig, replace emotion with the majoirity one
    # merge format 2:
    # usage, neutral, happiness, surprise, sadness, anger, disgust, fear, contempt, unknown, NF, pixels

    ferplus_col_id2emo_label, ferplus_emo2id = get_label2id_id2label(ferplus_emotion_labels)

    def argmax_idx(values: List[int]) -> int:
        # from operator import itemgetter
        # index, element = max(enumerate(items), key=itemgetter(1))
        return values.index(max(values))
    ferplus_entries = []


    with open(ferplus_csv_path,'r') as csvfile:
        ferplus_rows = csv.reader(csvfile, delimiter=',')
        for row in islice(ferplus_rows, 1, None):
            ferplus_entries.append(row)

    index = 0
    count_fer_ferplus_match = 0  # how many images with the same label (majority label for ferplus)
    ferplus_merged1_entries = [fer_orig_cols]  # first row is the header with the col names
    fer_orig_emo_counter = {emo: 0 for emo in fer_orig_emotion_labels}
    ferplus_majority_emo_counter = {emo: 0 for emo in ferplus_emotion_labels}
    with open(fer_orig_csv_path,'r') as csvfile:
        fer_rows = csv.reader(csvfile, delimiter=',')
        for fer_orig_row in islice(fer_rows, 1, None):
            # fer_orig_row: from this we take the pixels and usage (check that is same as usage in ferplus)
            # ferplus_row: from this we calculate the majority emotion
            ferplus_row = ferplus_entries[index]
            file_name = ferplus_row[FERP_COL_FILENAME].strip()
            if len(file_name) > 0: # todo: check why some rows are empty
                # image = str_to_image(fer_orig_row[FERO_COL_IMG])
                # image_path = os.path.join(base_folder, folder_names[fer_orig_row[FERO_COL_USAGE]], file_name)
                # image.save(image_path, compress_level=0)

                assert len(ferplus_row) == 2 + len(ferplus_emotion_labels), f"{len(ferplus_row)=} {ferplus_row=}" # 8 cols for emotion labels, 1 col for usage, 1 col for filename
                label_votes: List[int] = ferplus_row[2:]
                assert len(label_votes) == len(ferplus_emotion_labels), f"{len(label_votes)=} {len(ferplus_emotion_labels)=}"

                majority_label_id: int = argmax_idx(label_votes)
                usage_fer: str = fer_orig_row[FERO_COL_USAGE]
                usage_ferplus: str = ferplus_row[FERP_COL_USAGE]
                assert usage_fer == usage_ferplus

                fer_orig_emo_id: int = int(fer_orig_row[FERO_COL_EMO])
                fer_orig_emo_label: str = fer_orig_emotion_labels[fer_orig_emo_id]
                ferplus_majority_label: str = ferplus_emotion_labels[majority_label_id]
                if fer_orig_emo_label.lower()[:3] == ferplus_majority_label[:3]:
                    count_fer_ferplus_match += 1
                fer_orig_emo_counter[fer_orig_emo_label] += 1
                ferplus_majority_emo_counter[ferplus_majority_label] += 1

                # "emotion,pixels,Usage"
                new_row: List = [majority_label_id, fer_orig_row[FERO_COL_IMG], usage_fer]
                ferplus_merged1_entries.append(new_row)
            index += 1

    print(f"{count_fer_ferplus_match=} out of {len(ferplus_entries)=}")
    print(f"{fer_orig_emo_counter=}")
    print(f"{ferplus_majority_emo_counter=}")

    print(f"Saving merged file..")
    with open(out_merged_csv_path, 'w', newline='') as csvfile:
        csv_writer = csv.writer(csvfile, delimiter=',')
        csv_writer.writerows(ferplus_merged1_entries)

    print("Done...")

# https://discuss.pytorch.org/t/computing-the-mean-and-std-of-dataset/34949/13
def get_img_transforms(do_transforms: bool):
        # define transforms
    transforms_list = [
            v2.Resize((64,64)), # v2.Resize((48,48)),  #
            # v2.ToImage(),  # Convert to tensor, only needed if you had a PIL image
            transforms.ToTensor(), # v2.ToTensor(),
    ]
    if do_transforms:
        transforms_list.extend([
            v2.RandomHorizontalFlip(),
            # v2.RandomEqualize(), #  randomly equalizes the histogram of the given image.
            # v2.RandomRotation(degrees=(0, 180)),

            v2.RandomAffine(degrees=(30, 70), translate=(0.1, 0.3), ), # scale=(0.5, 0.75)

            # normalize,
            # ?rotate? transform = np.array([[np.cos(ang), -np.sin(ang)], [np.sin(ang), np.cos(ang)]])
            # skew_y transform = transform.dot(np.array([[1.0, skew_y], [0.0, 1.0]]))
            # skew_x transform = transform.dot(np.array([[1.0, 0.0], [skew_x, 1.0]]))
            # crop, scale, transform = transform.dot(np.diag([s_y, s_x]))
            # shift
            # v2.RandAugment(),

            # NB: normalization create strange artifacts in the img
            # v2.Normalize(mean=[0.5], std=[0.5]),
        ])
    transform = transforms.Compose(transforms_list)
    return transform

def data_loader_FER(
        data_dir,
        batch_size,
        random_seed=42,
        # valid_size=0.1,
        shuffle=True,
        test=False,
        test_dataset_pkl_path=None,
        train_dataset_pkl_path=None,
        valid_dataset_pkl_path=None,
        drop_last = True,
        do_transforms = True,
    ):
    assert test_dataset_pkl_path is not None and isinstance(test_dataset_pkl_path, str)
    assert train_dataset_pkl_path is not None and isinstance(train_dataset_pkl_path, str)
    assert valid_dataset_pkl_path is not None and isinstance(valid_dataset_pkl_path, str)
    # normalize = transforms.Normalize(
    #     mean=[0.4914, 0.4822, 0.4465],
    #     std=[0.2023, 0.1994, 0.2010],
    # )

    transform = get_img_transforms(do_transforms=do_transforms)
    assert transform is not None
    # do data transformations
    #  do not return more images (tr set count)
    #  but the data loader each time returns a randomly altered image
    #  first transforms: shift, angle, skew, ..uniform, scale x/y (random y or n), crop, fpliplr (random y/n)  : distort_img
    #  then resize to 64x64 (and possibly other stuff) : imgu.preproc_img
    #
    # normalization ?
    # (seems fer+ only normalizes within an image, not with same mean/std across the dataset)
    #
    # check the use of ROI box with the face in an image, in the label.csv file of fer+. How is calculated/obtained?
    #
    # "[33] Z. Yu and C. Zhang (2015)
    # Image based static facial expression recognition with multiple deep network learning.
    # In Proceedings of the 2015 ACM on International Conference on Multimodal Interaction,
    # ICMI ’15, pages 435{442, New York, NY, USA, 2015.
    # ACM."
    #
    #  Yu et al. (2015) "We adopted stochastic pooling [37] instead of max pooling
    # for its good performance given limited training data. Unlike
    # max pooling which chooses the maximum response, stochastic pooling randomly samples a response based on the probability distribution obtained by normalizing the responses."
    #
    #  Yu et al. (2015) "The network is trained using the adaptive subgradient
    # method [12] with a batch size of 128 examples."
    #
    # " the classification performance can be further improved if we
    # randomly perturb the input faces with additional transforms"
    # "The random
    # perturbation essentially generates additional unseen training
    # samples and therefore makes the network even more robust
    # to deviated and rotated faces
    #
    # A similar method is reported in [16] where the authors
    # generate perturbed training data by feeding their network
    # with randomly cropped and flipped 40×40 face images from
    # the original ones.
    # ("Static Facial Expression Recognition")
    #
    # " we consider a much more
    # comprehensive set of perturbations through the following
    # randomized affine image warping"
    #
    #
    # "During training we augment the data set on the fly, applying affine transforms similar to those in [33]. Such data
    # augmentation has been proven to improve the robustness of the model against translation, rotation and scaling.""
    # transformations: ? translation, rotation and scaling, cropping?
    # stretch images to 64x64
    # normalize images
    # todo: find in torch vision equivalent img transform as those done in fer+
    # data aumentation to rebalance labels
    # https://github.com/microsoft/FERPlus/blob/master/src/img_util.py
    # https://github.com/microsoft/FERPlus/blob/master/src/ferplus.py
    #
    # class FERPlusReader(object):
    #  def next_minibatch(self, batch_size):
    #  '''
    #         Return the next mini-batch, we do data augmentation during constructing each mini-batch.
    #         '''
    # distort_img
    #    distorted_image = imgu.distort_img(self.data[index][1],
        #    self.data[index][3],
        #    self.width,
        #    self.height,
        #    self.max_shift,
        #    self.max_scale,
        #    self.max_angle,
        #    self.max_skew,
        #    self.do_flip)
    # -- T_im = crop_img(img, roi, out_width, out_height, shift_x, shift_y, scale_x, scale_y, angle, sk_x, sk_y)
    # --  if flip and rnd.choice([True, False]): T_im = np.fliplr(T_im)
    # preproc_img
    #  final_image = imgu.preproc_img(distorted_image, A=self.A, A_pinv=self.A_pinv)
    #
    # _process_target
    #
    # https://github.com/microsoft/FERPlus/blob/master/src/train.py
    # in each epoch:
    # images, labels, current_batch_size = train_data_reader.next_minibatch(minibatch_size)
    #

    if test:
        if os.path.exists(test_dataset_pkl_path):
            test_dataset: FER2013Dataset = load_from_pickle(test_dataset_pkl_path)
        else:
            test_dataset = FER2013Dataset(
                root=data_dir,
                train=False,
                usage_split = "PublicTest",
                transform=transform,
            )
            save2pickle(test_dataset_pkl_path, test_dataset)
        print(f"{len(test_dataset)=}")
        data_loader: DataLoader = DataLoader(
            test_dataset, batch_size=batch_size, shuffle=shuffle,
        )

        return data_loader

    # load the dataset
    if os.path.exists(train_dataset_pkl_path):
        train_dataset: FER2013Dataset = load_from_pickle(train_dataset_pkl_path)
    else:
        train_dataset = FER2013Dataset(
            root=data_dir,
            train=True,
            usage_split = "Training",
            transform=transform,
        )
        save2pickle(train_dataset_pkl_path, train_dataset)
    print(f"{len(train_dataset)=}")

    if os.path.exists(valid_dataset_pkl_path):
        valid_dataset: FER2013Dataset = load_from_pickle(valid_dataset_pkl_path)
    else:
        valid_dataset = FER2013Dataset(
            root=data_dir,
            train=False,
            usage_split = "PrivateTest",
            transform=transform,
        )
        save2pickle(valid_dataset_pkl_path, valid_dataset)
    print(f"{len(valid_dataset)=}")

    def get_dataset_sampler(dataset):
        num_samples = len(dataset)
        indices = list(range(num_samples))
        # split = int(np.floor(valid_size * num_train))

        if shuffle:
            np.random.seed(random_seed)
            np.random.shuffle(indices)

            # train_idx, valid_idx = indices[split:], indices[:split]

            return SubsetRandomSampler(indices)

    train_loader: DataLoader = DataLoader(
        train_dataset, batch_size=batch_size, sampler=get_dataset_sampler(train_dataset), drop_last=drop_last)
    if drop_last:
        last_mb_size = len(train_dataset) % batch_size
        if last_mb_size != 0:
            print(f"Will drop the last mb_size: {last_mb_size=}, {len(train_dataset)=}, {batch_size=}, {drop_last=} ")

    valid_loader: DataLoader = DataLoader(
        valid_dataset, batch_size=batch_size, sampler=get_dataset_sampler(valid_dataset))

    return (train_loader, valid_loader)


def str_to_image(image_blob) -> Image:
    ''' Convert a string blob to an image object. '''
    image_data = str_to_np(image_blob)
    return Image.fromarray(image_data.reshape(48,48))

def str_to_np(image_blob) -> np.ndarray:
    ''' Convert a string blob to an np.asarray (linear 1D). '''
    image_string = image_blob.split(' ')
    image_data = np.asarray(image_string, dtype=np.uint8) # .reshape(48,48)
    return image_data

class FERPlusMergeDataset(VisionDataset):
    pass

    def get_sample_label_votes(self, index: int):
        # for each label, returns the number of votes for the sample at index
        pass

    def get_majority_class():
        pass

class FER2013Dataset(VisionDataset):

    # check prev loading of FER data (FERplus preprocessing)

    # datasets.CIFAR100
    # class CIFAR10(VisionDataset):
    """

    Args:
        root (str or ``pathlib.Path``): Root directory of dataset where directory
            ``cifar-10-batches-py`` exists or will be saved to if download is set to True.
        train (bool, optional): If True, creates dataset from training set, otherwise
            creates from test set.
        transform (callable, optional): A function/transform that takes in a PIL image
            and returns a transformed version. E.g, ``transforms.RandomCrop``
        target_transform (callable, optional): A function/transform that takes in the
            target and transforms it.
        download (bool, optional): If true, downloads the dataset from the internet and
            puts it in root directory. If dataset is already downloaded, it is not
            downloaded again.

    """

    def __init__(
        self,
        root: Union[str, Path],
        train: bool = True,
        transform: Optional[Callable] = None,
        target_transform: Optional[Callable] = None,
        # download: bool = False,
        usage_split: str = "PrivateTest",
    ) -> None:
        super().__init__(root, transform=transform, target_transform=target_transform)

        if train:
            assert "Training" == usage_split, f"invalid {usage_split=}, {train=}"
        else:
            assert usage_split in ["PrivateTest", "PublicTest"], f"invalid {usage_split=}, {train=}"
        self.train = train  # training set or test set

        self.data: List[np.ndarray] = []
        self.targets: List[int] = []
        fer_path = root

        # cols: emotion,pixels,Usage

        with open(fer_path) as f:
            total_lines = sum(1 for line in f)

        index = 0
        with open(fer_path,'r') as csvfile:
            fer_rows: csv._reader = csv.reader(csvfile, delimiter=',')
            with tqdm(total=total_lines) as pbar:
                for row in islice(fer_rows, 1, None):
                    usage_split_str = row[2]
                    emotion: int = int(row[0])
                    assert usage_split_str in ["Training", "PrivateTest", "PublicTest"], f"invalid {usage_split_str=}, {index=}, {emotion=}"
                    if usage_split == usage_split_str:
                        # image: Image = str_to_image(row[1])
                        image: np.ndarray = str_to_np(row[1])
                        assert len(image) == 48*48 # 2304=48*48
                        assert image.shape == (48*48 ,)
                        # print(f"{image.shape=}, {image.size=}") # image.shape=(2304,), image.size=2304
                        # image2: Image = str_to_image(row[1])
                        # image2.show()
                        # print(f"{np.min(image)=}, {np.max(image)=}, {image.dtype=}") # dtype('uint8')
                        # assert np.min(image) >= 0 and np.max(image) <= 255
                        # assert image.dtype == np.dtype('uint8')
                        # print(f"{type(emotion)=}")
                        # break
                        # Image.fromarray(image.reshape(48,48)).show()
                        self.data.append(image)
                        self.targets.append(emotion)
                        #
                        # return
                        #usage_split_train = True if usage_split_str == "Training" else False

                    index += 1
                    pbar.update(1)
            # Image.fromarray(self.data[0].reshape(48,48)).show()
        # todo: normalization by mean and std
        # https://discuss.pytorch.org/t/computing-the-mean-and-std-of-dataset/34949/14

        # self.data = np.vstack(self.data).reshape(-1, 3, 32, 32)
        # self.data = self.data.transpose((0, 2, 3, 1))  # convert to HWC


    def __getitem__(self, index: int) -> Tuple[torch.Tensor, int]:
        """
        Args:
            index (int): Index

        Returns:
            tuple: (image, target) where target is index of the target class.
        """
        # assert index < len(self.data), f"list index out of range {index=}, {len(self.data)=}"
        img_arr: np.ndarray
        target: int
        img_arr, target = self.data[index], self.targets[index]

        # doing this so that it is consistent with all other datasets
        # to return a PIL Image

        assert len(img_arr) == 48*48
        assert img_arr.shape == (48*48,)

        img: Image = Image.fromarray(img_arr.reshape(48,48))  #
        # removed reshape to avoid error TypeError: default_collate: batch must contain tensors, numpy arrays, numbers, dicts or lists; found <class 'PIL.Image.Image'>
        # img.show(title="pre-transform")

        if self.transform is not None:
            img_t: torch.Tensor = self.transform(img)
            # img.show(title="post-transform")
        else:
            img_t = pil_to_tensor(img)
        if self.target_transform is not None:
            target = self.target_transform(target)

        return img_t, target


    def __len__(self) -> int:
        return len(self.data)



class VGG13FerPlus(nn.Module):
    def __init__(self, num_classes, in_channels, fc_in_dim: int, # net_width:int=6,
                 ):
        # in_channels: 1 for greyscale, 3 for rgb

        super(VGG13FerPlus, self).__init__()
        self.layer1 = nn.Sequential(
            nn.Conv2d(in_channels, 64, kernel_size=3, stride=1, padding=1),
            nn.BatchNorm2d(64),
            nn.ReLU())
        self.layer2 = nn.Sequential(
            nn.Conv2d(64, 64, kernel_size=3, stride=1, padding=1),
            nn.BatchNorm2d(64),
            nn.ReLU(),
            nn.MaxPool2d(kernel_size = 2, stride = 2))

        self.layer3 = nn.Sequential(
            nn.Dropout(p=0.25), # During training, randomly zeroes some of the elements of the input tensor with probability p
            nn.Conv2d(64, 128, kernel_size=3, stride=1, padding=1),
            nn.BatchNorm2d(128),
            nn.ReLU())
        self.layer4 = nn.Sequential(
            nn.Conv2d(128, 128, kernel_size=3, stride=1, padding=1),
            nn.BatchNorm2d(128),
            nn.ReLU(),
            nn.MaxPool2d(kernel_size = 2, stride = 2))

        self.layer5 = nn.Sequential(
            nn.Dropout(0.25),
            nn.Conv2d(128, 256, kernel_size=3, stride=1, padding=1),
            nn.BatchNorm2d(256),
            nn.ReLU())
        self.layer6 = nn.Sequential(
            nn.Conv2d(256, 256, kernel_size=3, stride=1, padding=1),
            nn.BatchNorm2d(256),
            nn.ReLU())
        self.layer7 = nn.Sequential(
            nn.Conv2d(256, 256, kernel_size=3, stride=1, padding=1),
            nn.BatchNorm2d(256),
            nn.ReLU(),
            nn.MaxPool2d(kernel_size = 2, stride = 2))

        self.layer8 = nn.Sequential(
            nn.Dropout(0.25),
            nn.Conv2d(256, 256, kernel_size=3, stride=1, padding=1),
            nn.BatchNorm2d(256),
            nn.ReLU())
        self.layer9 = nn.Sequential(
            nn.Conv2d(256, 256, kernel_size=3, stride=1, padding=1),
            nn.BatchNorm2d(256),
            nn.ReLU())
        self.layer10 = nn.Sequential(
            nn.Conv2d(256, 256, kernel_size=3, stride=1, padding=1),
            nn.BatchNorm2d(256),
            nn.ReLU(),
            nn.MaxPool2d(kernel_size = 2, stride = 2))

        self.fc = nn.Sequential(
            nn.Dropout(0.5),
            nn.Linear(fc_in_dim, 1024),  #  VGG13, 1ch: 9x256 VGG13, 3ch (rgb):196*256, 1024  VGG16: 7*7*512, 4096 (196 = 7x7x4)
            nn.ReLU())
        self.fc1 = nn.Sequential(
            nn.Dropout(0.5),
            nn.Linear(1024, 1024),
            nn.ReLU())
        self.fc2= nn.Sequential(
            nn.Linear(1024, num_classes))

    def forward(self, x):
        out = self.layer1(x)
        out = self.layer2(out)
        out = self.layer3(out)
        out = self.layer4(out)
        out = self.layer5(out)
        out = self.layer6(out)
        out = self.layer7(out)
        out = self.layer8(out)
        out = self.layer9(out)
        out = self.layer10(out)
        out = out.reshape(out.size(0), -1)
        out = self.fc(out)
        out = self.fc1(out)
        out = self.fc2(out)
        return out

# class VGG16(nn.Module):
#     def __init__(self, num_classes=10):
#         super(VGG16, self).__init__()
#         self.layer1 = nn.Sequential(
#             nn.Conv2d(3, 64, kernel_size=3, stride=1, padding=1),
#             nn.BatchNorm2d(64),
#             nn.ReLU())
#         self.layer2 = nn.Sequential(
#             nn.Conv2d(64, 64, kernel_size=3, stride=1, padding=1),
#             nn.BatchNorm2d(64),
#             nn.ReLU(),
#             nn.MaxPool2d(kernel_size = 2, stride = 2))
#         self.layer3 = nn.Sequential(
#             nn.Conv2d(64, 128, kernel_size=3, stride=1, padding=1),
#             nn.BatchNorm2d(128),
#             nn.ReLU())
#         self.layer4 = nn.Sequential(
#             nn.Conv2d(128, 128, kernel_size=3, stride=1, padding=1),
#             nn.BatchNorm2d(128),
#             nn.ReLU(),
#             nn.MaxPool2d(kernel_size = 2, stride = 2))
#         self.layer5 = nn.Sequential(
#             nn.Conv2d(128, 256, kernel_size=3, stride=1, padding=1),
#             nn.BatchNorm2d(256),
#             nn.ReLU())
#         self.layer6 = nn.Sequential(
#             nn.Conv2d(256, 256, kernel_size=3, stride=1, padding=1),
#             nn.BatchNorm2d(256),
#             nn.ReLU())
#         self.layer7 = nn.Sequential(
#             nn.Conv2d(256, 256, kernel_size=3, stride=1, padding=1),
#             nn.BatchNorm2d(256),
#             nn.ReLU(),
#             nn.MaxPool2d(kernel_size = 2, stride = 2))
#         self.layer8 = nn.Sequential(
#             nn.Conv2d(256, 512, kernel_size=3, stride=1, padding=1),
#             nn.BatchNorm2d(512),
#             nn.ReLU())
#         self.layer9 = nn.Sequential(
#             nn.Conv2d(512, 512, kernel_size=3, stride=1, padding=1),
#             nn.BatchNorm2d(512),
#             nn.ReLU())
#         self.layer10 = nn.Sequential(
#             nn.Conv2d(512, 512, kernel_size=3, stride=1, padding=1),
#             nn.BatchNorm2d(512),
#             nn.ReLU(),
#             nn.MaxPool2d(kernel_size = 2, stride = 2))
#         self.layer11 = nn.Sequential(
#             nn.Conv2d(512, 512, kernel_size=3, stride=1, padding=1),
#             nn.BatchNorm2d(512),
#             nn.ReLU())
#         self.layer12 = nn.Sequential(
#             nn.Conv2d(512, 512, kernel_size=3, stride=1, padding=1),
#             nn.BatchNorm2d(512),
#             nn.ReLU())
#         self.layer13 = nn.Sequential(
#             nn.Conv2d(512, 512, kernel_size=3, stride=1, padding=1),
#             nn.BatchNorm2d(512),
#             nn.ReLU(),
#             nn.MaxPool2d(kernel_size = 2, stride = 2))
#         self.fc = nn.Sequential(
#             nn.Dropout(0.5),
#             nn.Linear(7*7*512, 4096),
#             nn.ReLU())
#         self.fc1 = nn.Sequential(
#             nn.Dropout(0.5),
#             nn.Linear(4096, 4096),
#             nn.ReLU())
#         self.fc2= nn.Sequential(
#             nn.Linear(4096, num_classes))

#     def forward(self, x):
#         out = self.layer1(x)
#         out = self.layer2(out)
#         out = self.layer3(out)
#         out = self.layer4(out)
#         out = self.layer5(out)
#         out = self.layer6(out)
#         out = self.layer7(out)
#         out = self.layer8(out)
#         out = self.layer9(out)
#         out = self.layer10(out)
#         out = self.layer11(out)
#         out = self.layer12(out)
#         out = self.layer13(out)
#         out = out.reshape(out.size(0), -1)
#         out = self.fc(out)
#         out = self.fc1(out)
#         out = self.fc2(out)
#         return out

def check_compare_datasets(fer_csv_path):
    batch_size_1 = 32 # 64
    train_loader_FER: DataLoader
    train_loader_FER, valid_loader_FER = data_loader_FER(data_dir=fer_csv_path,
                                            batch_size=batch_size_1,
                                            # valid_size=0.99,
                                            )

    test_loader_FER = data_loader_FER(data_dir=fer_csv_path,
                                batch_size=batch_size_1,
                                test=True)

    # train_loader_cifar, valid_loader_cifar = data_loader_cifar(data_dir='./data',
    #                                         batch_size=batch_size_1, valid_size=0.99)

    # test_loader_cifar = data_loader_cifar(data_dir='./data',
    #                             batch_size=batch_size_1,
    #                             test=True)


    # Display image and label.
    for dataloader in [train_loader_FER, #train_loader_cifar
                       ]:
        train_features, train_labels = next(iter(train_loader_FER))
        print(f"Feature batch shape: {train_features.size()}")
        print(f"Labels batch shape: {train_labels.size()}")
        img = train_features[0].squeeze()
        label = train_labels[0]
        # plt.imshow(img, cmap="gray")
        # plt.show()
        print(f"Label: {label}")
    # assert type(train_loader_FER[0]) == type(test_loader_cifar[0]), f"{type(train_loader_FER[0])=} {type(test_loader_cifar[0])=}"


def get_model_config() -> Dict[str, Any]:
    VGG16_Config = {"num_classes": 100, "in_channels": 3, "fc_in_dim": 7*7*512, "batch_size": 32}

    # "fc_in_dim": 9*256, # for 48x48
    # "fc_in_dim": 16*256, # for 64x64
    VGG13FER_Config_48size = {"num_classes": len(ferplus_emotion_labels), "in_img_edge_size": 48, "in_channels": 1, "fc_in_dim": 9*256, "batch_size": 952, "lr": 0.05} # 640
    VGG13FER_Config_64size = {"num_classes": len(ferplus_emotion_labels), "in_img_edge_size": 64, "in_channels": 1, "fc_in_dim": 16*256, "batch_size": 460, "lr": 0.05} # "batch_size": 460
    config = VGG13FER_Config_64size # VGG16_Config
    num_classes = config["num_classes"]
    print(f"{num_classes=}")
    in_channels = config["in_channels"]
    fc_in_dim = config["fc_in_dim"]

    batch_size = config["batch_size"]

    # batch_size_2 = 3 # 16
    learning_rate = config["lr"] # 0.005

    return config

def train_test_net(fer_csv_path):

    # Device configuration
    device = torch.device('cuda' if torch.cuda.is_available() else 'cpu')
    print(f"{device=}")
    # fc_in_dim: VGG13, 1ch: 9*256 VGG13, 3ch (rgb):196*256, 1024  VGG16: 7*7*512, 4096 (196 = 7x7x4)

    config = get_model_config()
    num_epochs = 100 # 20
    do_validation = True
    do_testing = True

    train_loader: DataLoader
    valid_loader: DataLoader
    train_loader, valid_loader = data_loader_FER(
        data_dir=fer_csv_path,
        batch_size=config["batch_size"],
        # valid_size=0.2,
        test_dataset_pkl_path="test_dataset.pkl",
        train_dataset_pkl_path="train_dataset.pkl",
        valid_dataset_pkl_path="valid_dataset.pkl",
    )

    test_loader: DataLoader = data_loader_FER(
        data_dir=fer_csv_path,
        batch_size=config["batch_size"],
        test=True,
        test_dataset_pkl_path="test_dataset.pkl",
        train_dataset_pkl_path="train_dataset.pkl",
        valid_dataset_pkl_path="valid_dataset.pkl",
    )
    config["in_img_edge_size"]


    model = VGG13FerPlus(num_classes=config["num_classes"], in_channels=config["in_channels"], fc_in_dim=config["fc_in_dim"]).to(device)
    summary(model, input_size=(config["in_channels"], config["in_img_edge_size"], config["in_img_edge_size"]), batch_size=config["batch_size"])
    # , batch_size=-1, device="cuda"

    # Loss and optimizer
    criterion = nn.CrossEntropyLoss()
    optimizer = torch.optim.SGD(model.parameters(), lr=config["lr"], weight_decay = 0.005, momentum = 0.9)
    # NB: this is not scheduled according to number of steps / epocs (check why/if needs to)


    # Train the model
    total_step = len(train_loader)

    def get_correct_count_in_batch(outputs: torch.Tensor, labels: torch.Tensor) -> int:
        _, predicted = torch.max(outputs.data, 1)
        in_batch_count = labels.size(0)
        return int((predicted == labels).sum().item())


    plt.ion()
    plt.axis()  # [0, 10, 0, 1] [xmin, xmax, ymin, ymax]
    # plot tr loss at each step, vl loss at each epoch
    # x axis is
    abs_step = 0
    tr_steps_accuracies = []
    tr_steps_loss = []
    val_losses = []

    print(f"starting epochs..")
    for epoch in range(num_epochs):
        correct_tr_epoch = 0
        correct_tr_step = 0
        total = 0
        for step_num, (images, labels) in enumerate(tqdm(train_loader)):

            # Move tensors to the configured device
            images = images.to(device)
            labels = labels.to(device)

            # Forward pass
            outputs = model(images)
            # accuracy calculation: check if we need to move the data back to cpu device
            correct_tr_step = get_correct_count_in_batch(outputs, labels)
            in_batch_count = labels.size(0)
            total += in_batch_count
            correct_tr_epoch += correct_tr_step
            accuracy_step = 100 * correct_tr_step / len(images)
            curr_abs_step_num = (epoch * step_num) + step_num

            tr_steps_accuracies.append(accuracy_step)

            # print(f"{images.size()=}, {outputs.size()=}, {labels.size()=},")
            # type(images)=<class 'torch.Tensor'>, type(outputs)=<class 'torch.Tensor'>, type(labels)=<class 'torch.Tensor'>,
            # mages.size()=torch.Size([460, 1, 64, 64]), outputs.size()=torch.Size([460, 10]), labels.size()=torch.Size([460]),
            _, pred = torch.max(outputs,1) # Returns a namedtuple ``(values, indices)`` where ``values`` is the maximum
            # value of each row of the :attr:`input` tensor in the given dimension
            # :attr:`dim`. And ``indices`` is the index location of each maximum value found
            # (argmax).
            idxs_mask = ((pred == labels) == False).nonzero()
            idxs_mask = torch.squeeze(idxs_mask)
            # print(f"{idxs_mask.size()=}, {torch.squeeze(idxs_mask).size()=}")
            # print(f"{torch.squeeze(idxs_mask)[0]=}")
            # print(f"{torch.squeeze(idxs_mask)[1]=}")
            # print(f"{torch.max(torch.squeeze(idxs_mask))=}")
            misclassified_imgs = torch.index_select(images, 0, idxs_mask)
            misclassified_true_labels = torch.index_select(labels, 0, idxs_mask)
            misclassified_preds = torch.index_select(pred, 0, idxs_mask)
            viz_misclassified_imgs(misclassified_imgs, misclassified_true_labels, misclassified_preds)
            # return

            # idxs_mask.size()=torch.Size([310, 1])
            loss: torch.Tensor = criterion(outputs, labels)
            tr_steps_loss.append(loss.item())

            total += labels.size(0)
            correct_tr_epoch += get_correct_count_in_batch(outputs, labels)

            # print(f"{loss.shape=} {loss.size=}  {loss.item()=} {loss=}")
            # oss.shape=torch.Size([]) loss.size=<built-in method size of Tensor object at 0x000002454DADFCF0>  loss.item()=2.3087422847747803 loss=tensor(2.3087, device='cuda:0', grad_fn=<NllLossBackward0>)

            # Backward and optimize
            optimizer.zero_grad()
            loss.backward()
            optimizer.step()

        print ('Epoch [{}/{}], Step [{}/{}], Tr. Loss: {:.4f}'
                    .format(epoch+1, num_epochs, step_num+1, total_step, loss.item()))
        print(f'Accuracy of the network on the {len(train_loader.dataset)} training images: {100 * correct_tr_epoch / total:.4f} %')

        # Validation
        print(f"starting validation..")
        if do_validation:
            with torch.no_grad():
                correct = 0
                total = 0
                for images, labels in valid_loader:  # tqdm(valid_loader)
                    images = images.to(device)
                    labels = labels.to(device)
                    outputs = model(images)

                    # check misclassified imgs, for later visualization
                    print(f"{type(images)=}, {type(outputs)=}, {type(labels)=},")
                    val_loss: torch.Tensor = criterion(outputs, labels)
                    val_losses.append(val_loss.item())

                    total += labels.size(0)
                    correct += get_correct_count_in_batch(outputs, labels)
                    del images, labels, outputs
                print(f'Val loss: {val_loss.item():.4f}. Accuracy of the network on the {len(valid_loader.dataset)} validation images: {100 * correct / total:.4f} %')
                # print('Accuracy of the network on the {} validation images: {} %'.format(5000, 100 * correct / total))
                # todo: print one (or a figure with multiple) image correctly classified and one not


    print(f"starting test..")
    if do_testing:
        with torch.no_grad():
            correct = 0
            total = 0
            for images, labels in tqdm(test_loader):
                images = images.to(device)
                labels = labels.to(device)
                outputs = model(images)

                # for misclassified imgs show: img, predicted, actual

                test_loss: torch.Tensor = criterion(outputs, labels)
                # test_losses.append(val_loss.item())

                _, predicted = torch.max(outputs.data, 1)
                total += labels.size(0)
                correct += (predicted == labels).sum().item()
                del images, labels, outputs
            print(f'Test loss: {test_loss.item():.4f}. Accuracy of the network on the {len(test_loader.dataset)} test images: {100 * correct / total:.4f} %')
            # print('Accuracy of the network on the {} test images: {} %'.format(10000, 100 * correct / total))
    print(f"Done.")

def print_dataset_stats(fer_path: str, labels: List[str]):
    emo_counter = {emo: 0 for emo in labels}
    with open(fer_path,'r') as csvfile:
        fer_rows = csv.reader(csvfile, delimiter=',')
        for fer_orig_row in islice(fer_rows, 1, None):
            # fer_orig_row: from this we take the pixels and usage (check that is same as usage in ferplus)
            # ferplus_row: from this we calculate the majority emotion

            fer_orig_emo_id: int = int(fer_orig_row[FERO_COL_EMO])
            fer_orig_emo_label: str = fer_orig_emotion_labels[fer_orig_emo_id]
            emo_counter[fer_orig_emo_label] += 1

    print(f"{emo_counter=}")

# count_fer_ferplus_match=22360 out of len(ferplus_entries)=35887
# fer_orig_emo_counter={'Angry': 4925, 'Disgust': 546, 'Fear': 5103, 'Happy': 8953, 'Sad': 6038, 'Surprise': 3989, 'Neutral': 6160}
# ferplus_majority_emo_counter={'neutral': 12906, 'happiness': 9355, 'surprise': 4462, 'sadness': 4371, 'anger': 3111, 'disgust': 248, 'fear': 819, 'contempt': 216, 'unknown': 222, 'NF': 4}



def print_model_summary():
    config = get_model_config()
    model = VGG13FerPlus(num_classes=config["num_classes"], in_channels=config["in_channels"], fc_in_dim=config["fc_in_dim"]) # .to(device)
    summary(
        model, input_size=(config["in_channels"], config["in_img_edge_size"], config["in_img_edge_size"]), batch_size=config["batch_size"],
        device="cpu"
    )

    # VGG13FER_Config_64size
    # num_classes=10
    # ----------------------------------------------------------------
    #         Layer (type)               Output Shape         Param #
    # ================================================================
    #             Conv2d-1          [460, 64, 64, 64]             640
    #        BatchNorm2d-2          [460, 64, 64, 64]             128
    #               ReLU-3          [460, 64, 64, 64]               0
    #             Conv2d-4          [460, 64, 64, 64]          36,928
    #        BatchNorm2d-5          [460, 64, 64, 64]             128
    #               ReLU-6          [460, 64, 64, 64]               0
    #          MaxPool2d-7          [460, 64, 32, 32]               0
    #            Dropout-8          [460, 64, 32, 32]               0
    #             Conv2d-9         [460, 128, 32, 32]          73,856
    #       BatchNorm2d-10         [460, 128, 32, 32]             256
    #              ReLU-11         [460, 128, 32, 32]               0
    #            Conv2d-12         [460, 128, 32, 32]         147,584
    #       BatchNorm2d-13         [460, 128, 32, 32]             256
    #              ReLU-14         [460, 128, 32, 32]               0
    #         MaxPool2d-15         [460, 128, 16, 16]               0
    #           Dropout-16         [460, 128, 16, 16]               0
    #            Conv2d-17         [460, 256, 16, 16]         295,168
    #       BatchNorm2d-18         [460, 256, 16, 16]             512
    #              ReLU-19         [460, 256, 16, 16]               0
    #            Conv2d-20         [460, 256, 16, 16]         590,080
    #       BatchNorm2d-21         [460, 256, 16, 16]             512
    #              ReLU-22         [460, 256, 16, 16]               0
    #            Conv2d-23         [460, 256, 16, 16]         590,080
    #       BatchNorm2d-24         [460, 256, 16, 16]             512
    #              ReLU-25         [460, 256, 16, 16]               0
    #         MaxPool2d-26           [460, 256, 8, 8]               0
    #           Dropout-27           [460, 256, 8, 8]               0
    #            Conv2d-28           [460, 256, 8, 8]         590,080
    #       BatchNorm2d-29           [460, 256, 8, 8]             512
    #              ReLU-30           [460, 256, 8, 8]               0
    #            Conv2d-31           [460, 256, 8, 8]         590,080
    #       BatchNorm2d-32           [460, 256, 8, 8]             512
    #              ReLU-33           [460, 256, 8, 8]               0
    #            Conv2d-34           [460, 256, 8, 8]         590,080
    #       BatchNorm2d-35           [460, 256, 8, 8]             512
    #              ReLU-36           [460, 256, 8, 8]               0
    #         MaxPool2d-37           [460, 256, 4, 4]               0
    #           Dropout-38                [460, 4096]               0
    #            Linear-39                [460, 1024]       4,195,328
    #              ReLU-40                [460, 1024]               0
    #           Dropout-41                [460, 1024]               0
    #            Linear-42                [460, 1024]       1,049,600
    #              ReLU-43                [460, 1024]               0
    #            Linear-44                  [460, 10]          10,250
    # ================================================================
    # Total params: 8,763,594
    # Trainable params: 8,763,594
    # Non-trainable params: 0
    # ----------------------------------------------------------------
    # Input size (MB): 7.19
    # Forward/backward pass size (MB): 11719.25
    # Params size (MB): 33.43
    # Estimated Total Size (MB): 11759.87
    # ----------------------------------------------------------------
    #

# step_total, epoch_step, epoch, tr_loss, val_loss, val_acc/error_rate

def viz_images():
    # from torchvision.transforms.functional import to_pil_image # (pic, mode=None)
    ferplus_id2emo, ferplus_emo2id = get_label2id_id2label(ferplus_emotion_labels)
    print(f"{ferplus_id2emo=}")
    config = get_model_config()
    transform = get_img_transforms(do_transforms=True)
    test_dataset_no_transform = FER2013Dataset(
                root=fer_csv_path,
                train=False,
                usage_split = "PublicTest",
                transform=None,
            )
    test_dataset_with_transform = FER2013Dataset(
                root=fer_csv_path,
                train=False,
                usage_split = "PublicTest",
                transform=transform,
            )

    def show_img(dataset, img_idx: int, transform=None):
        img: torch.Tensor

        img, target = dataset[img_idx]
        print(f"{img_idx=} {target=} {ferplus_id2emo[target]=}")
        print(f"{len(dataset)=}")
        print(f"{type(img)=}")  # torch.Tensor
        print(f"{img.size()=}, {img.shape=}")
        # print(f"{type(images[img_idx])=} {images[img_idx].size()=}, {images[img_idx].shape=}")
        # img: Image = to_pil_image(img)
        img_pil = to_pil_image(img)
        img_pil.show()
        if transform is not None:
            img_pil = to_pil_image(transform(img_pil))
            print("Showing transformed img:")
            img_pil.show()
        else:
            print("transform is None")


    random_idx = 17 # 14 # 7 # 5 # 4 # 3
    show_img(test_dataset_no_transform, random_idx, transform=transform)  # Size([1, 48, 48])
    show_img(test_dataset_with_transform, random_idx)  # Size([1, 64, 64])

    data_loader: DataLoader = DataLoader(
        test_dataset_with_transform, batch_size=config["batch_size"], shuffle=False,
    )

    images: torch.Tensor
    for step_num, (images, labels) in enumerate(data_loader):
        print(f"{type(images)=}") # torch.Tensortorch.Tensor
        print(f"{images.size()=}, {images.shape=}")
        print(f"{type(images[random_idx])=} {images[random_idx].size()=}, {images[random_idx].shape=}")
        img: Image = to_pil_image(images[random_idx])
        # img.show()
        # todo: disable random sampler
        # bool var to disable transforms
        # show img at same index with and without transforms
        break
    return

    test_loader: DataLoader = data_loader_FER(
        data_dir=fer_csv_path,
        batch_size=config["batch_size"],
        test=True,
        test_dataset_pkl_path="test_dataset.pkl",
        train_dataset_pkl_path="train_dataset.pkl",
        valid_dataset_pkl_path="valid_dataset.pkl",
        shuffle=False,
        do_transforms=False,
    )

    images: torch.Tensor
    for step_num, (images, labels) in enumerate(test_loader):
        print(f"{type(images)=}")
        print(f"{images.size()=}, {images.shape=}")
        print(f"{images[random_idx].size()=}, {images[random_idx].shape=}")
        img: Image = to_pil_image(images[random_idx])
        img.show()
        # todo: disable random sampler
        # bool var to disable transforms
        # show img at same index with and without transforms
        break

    # inference mode: use a model to classify an image, show prediction and img (and emoji)
    # figure for multiple images, and label with prediction of each

def viz_misclassified_imgs(misclassified_imgs: torch.Tensor, misclassified_true_labels: torch.Tensor, misclassified_preds: torch.Tensor):
    plt.ioff()
    plt.figure()
    plt.tick_params(
        axis='y',          # changes apply to the x-axis
        which='both',      # both major and minor ticks are affected
        bottom=False,      # ticks along the bottom edge are off
        top=False,         # ticks along the top edge are off
        labelbottom=False
    ) # labels along the bottom edge are off
    plt.xticks([], [])
    # for axis object
    # ---------------
    # from Anakhand May 5 at 13:08
    # for major ticks


    #subplot(r,c) provide the no. of rows and columns
    f, axarr = plt.subplots(6,1)
    ferplus_id2emo, ferplus_emo2id = get_label2id_id2label(ferplus_emotion_labels)

    for figure_slot_idx in range(6):
        subplot: matplotlib.axes._axes.Axes = axarr[figure_slot_idx]
        img = misclassified_imgs[figure_slot_idx]
        true_labelid = misclassified_true_labels[figure_slot_idx]
        pred_label_id = misclassified_preds[figure_slot_idx]
        add_img_to_subplot(img, true_labelid, img_idx=-1,  subplot=subplot, ferplus_id2emo=ferplus_id2emo, pred_label_id=pred_label_id)  # Size([1, 48, 48])

    plt.show()

def add_img_to_subplot(
        img: torch.Tensor,
        true_labelid,
        img_idx: int,
        subplot: matplotlib.axes._axes.Axes,
        ferplus_id2emo,
        pred_label_id=None,
        transform=None
    ):

        # img, target = dataset[img_idx]

        # print(f"{img_idx=} {target=} {ferplus_id2emo[target]=}")
        # print(f"{len(dataset)=}")
        # print(f"{type(img)=}")  # torch.Tensor
        # print(f"{img.size()=}, {img.shape=}")
        # print(f"{type(images[img_idx])=} {images[img_idx].size()=}, {images[img_idx].shape=}")
        # img: Image = to_pil_image(img)
        img_pil = to_pil_image(img)
        # img_pil.show()
        if transform is not None:
            img_pil = to_pil_image(transform(img_pil))
            # print("Showing transformed img:")
            # img_pil.show()
        else:
            # print("transform is None")
            pass

        emo_true = ferplus_id2emo[true_labelid.item()]

        subplot_title: str = emo_true
        if pred_label_id is not None:
            emo_pred = ferplus_id2emo[pred_label_id.item()]
            subplot_title = f"pred: {emo_pred} | true: {subplot_title} "

        subplot.set_xticks([])
        subplot.set_yticks([])
        # for minor ticks
        subplot.set_xticks([], minor=True)
        subplot.xaxis.set_ticks_position('none')
        # subplot.title.set_text(emo_label) # , loc='right'
        subplot.set_title(subplot_title, loc='left')
        # subplot.yaxis.set_label_position('right')
        # subplot.set_ylabel(f"{emo_true} ({img_idx})", rotation=90)
        subplot.imshow(X=img_pil, cmap="gray", vmin=0, vmax=255)


def viz_img_transforms():

    # todo
    # add dataset name in the legend
    # (add descr property in dataset class)

    plt.figure()
    plt.tick_params(
        axis='y',          # changes apply to the x-axis
        which='both',      # both major and minor ticks are affected
        bottom=False,      # ticks along the bottom edge are off
        top=False,         # ticks along the top edge are off
        labelbottom=False
    ) # labels along the bottom edge are off
    plt.xticks([], [])
    # for axis object
    # ---------------
    # from Anakhand May 5 at 13:08
    # for major ticks


    #subplot(r,c) provide the no. of rows and columns
    f, axarr = plt.subplots(6,2)

    # captions: original, transform1, trasnform2, ..
    # todo: get the images from the dataloaders


    # # (pic, mode=None)
    ferplus_id2emo, ferplus_emo2id = get_label2id_id2label(ferplus_emotion_labels)
    print(f"{ferplus_id2emo=}")
    config = get_model_config()
    transform = get_img_transforms(do_transforms=True)
    test_dataset_no_transform = FER2013Dataset(
                root=fer_csv_path,
                train=False,
                usage_split = "PublicTest",
                transform=None,
            )
    test_dataset_with_transform = FER2013Dataset(
                root=fer_csv_path,
                train=False,
                usage_split = "PublicTest",
                transform=transform,
            )



    for img_idx, figure_slot_idx in zip([24, 19, 25, 26, 22, 23], range(6)):
        # 3, 4, 5, 7, 14, 19, 22, 23, 24, 25, 26,

        # random_idx = 17 # 14 # 7 # 5 # 4 # 3
        # figure_slot_idx = 0 # 0-3
        add_img_to_subplot(test_dataset_no_transform, img_idx=img_idx,  col_idx=0, figure_slot_idx=figure_slot_idx, axarr=axarr, ferplus_id2emo=ferplus_id2emo, transform=transform)  # Size([1, 48, 48])
        add_img_to_subplot(test_dataset_with_transform, img_idx=img_idx, col_idx=1, axarr=axarr, ferplus_id2emo=ferplus_id2emo, igure_slot_idx=figure_slot_idx, )  # Size([1, 64, 64])

    #todo: during validation and test, keep track of misclassified images, and diplay those

    # use the created array to output your multiple images. In this case I have stacked 4 images vertically
    # axarr[0].imshow(X=v_slice[0])  # For displaying a grayscale image, set up the colormapping using the parameters cmap='gray', vmin=0, vmax=255.
    # axarr[1].imshow(X=v_slice[1])  # Xarray-like or PIL image
    # axarr[2].imshow(X=v_slice[2])
    # axarr[3].imshow(X=v_slice[3])
    plt.show()

def main():
    # check_compare_datasets()
    # fer_orig_csv_path: str = "C:\\Users\\mauro\\Downloads\\FERCNN\\fer2013\\fer2013\\fer2013.csv"
    # merge_fer_ferplus(
    #     fer_orig_csv_path=fer_orig_csv_path,
    #     ferplus_csv_path="C:\\Users\\mauro\\Downloads\\FERCNN\\FERPlus-master\\FERPlus-master\\fer2013new.csv",
    #     out_merged_csv_path=get_fer_csv_dataset_path(),
    # )

    # print_model_summary()
    # viz_images()

    # viz_img_transforms()
    fer_csv_path = get_fer_csv_dataset_path()
    train_test_net(fer_csv_path)

if __name__ == "__main__":
    main()
```
